```python
# importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# load the dataset
data = pd.read_csv("hiring.csv")
data
```

```python
# check the shape
data.shape
```

```python
# check summary stats
data.describe()
```

```python
# checking info
data.info()
```

```python
# checking missing values
data.isnull().sum()
```

```python
# filling in missing values using fillna
data.fillna({'experience' : data['experience'].mode()[0]}, inplace=True)
data.fillna({'test_score(out of 10)' : data['test_score(out of 10)'].median()}, inplace=True)
```

```python
# check null values again
data.isnull().sum()
```

```python
# changing column name
data.rename(columns={'test_score(out of 10)' : 'test_score', 'interview_score(out of 10)' : 'interview_score', 'salary($)': 'salary'}, inplace=True)
```

```python
data.head()
```

```python
# preprocessing using get dummies -- converting experience to numerical values
dummy = pd.get_dummies(data[['experience']], drop_first=True, dtype=int)
dummy
```

```python
# merging
df_merged = pd.concat([data, dummy], axis = 1)
df_merged
```
