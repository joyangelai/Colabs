```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
# load dataset
data = pd.read_csv("/homeprices (1).csv")
data
```

```python
# get summary stats
data.describe()
```

```python
# check shape
data.shape
```

```python
# check info
data.info()
```

```python
# filling in missing values
data.isnull().sum()
```

```python
# using median
bedm = data['bedrooms'].median()
bedm
```

```python
# filling in missing values
data.fillna({'bedrooms' : bedm}, inplace=True)
```

```python
# checking for null values again
data.isnull().sum()
```

```python
# divide data into X and y
X = data.drop("price", axis=1)
y = data['price']
```

```python
# scale the model
from sklearn.preprocessing import StandardScaler

scl = StandardScaler()
x_scale = scl.fit_transform(X)
x_scale
```

```python
# init the model
mlr = linear_model.LinearRegression()
mlr

# train the model
mlr.fit(x_scale,y)
```

```python
# check coefficient
mlr.coef_
```

```python
# check intercept
mlr.intercept_  # --> this is the same as b0
```
