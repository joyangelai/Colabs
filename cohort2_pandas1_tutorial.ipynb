```python
#Import pandas and numpy
import pandas as pd # for data manipulation and analysis
import numpy as np # used for mathematical computation # built on top of pandas
```

Data stractures used in pandas
1. pandas series
2. pandas data Frame

```python
#Create a  num and str list and convert it to a series
a = ["car","boat", "bike"]  #list
series1 = pd.Series(a)
print(series1)
```

```python
#create a list and convert it to a dataframe
a = [("car",100),("boat",70), ("bike",40)]
df = pd.DataFrame(a, columns=("item","price"))
print(df)
```

```python
#saving the dataframe above to csv file
df.to_csv("item.csv", index= None)
```
