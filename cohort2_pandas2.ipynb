```python
#Create a  num and str list and convert it to a series
a = ["car", "boat", "bike"] #list

series1 = pd.Series(a)
print(series1)
print(type(series1))
```

```python
#create a list and convert it to a dataframe
a = [("car", 100), ("boat", 70), ("bike", 40)]

df = pd.DataFrame(a, columns=("item", "price"))
print(df)
type(df)
```

## Subsetting Columns

This is the act of selecting  a few columns from the entire dataframe.

```python
#Selecting a single colung using [] make it a series
df_dscount = dataDf['Discount']
print(type(df_dscount))
df_dscount.head()
```

```python
#selcting single column discount and making it a dataframe, use [[]]
df_dscount = dataDf[['Discount']]
print(type(df_dscount))
df_dscount.head()
```
