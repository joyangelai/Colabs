## Subsetting Columns

This is the act of selecting  a few columns from the entire dataframe.

```python
# Selecting a single column using []
data_discount = data['Discount']
print(type(data_discount))
data_discount.head()
```

```python
#selecting single column discount and making it a data frame, use [[]]
data_discount = data[['Discount']]
print(type(data_discount))
data_discount.head()
```

```python
#another way of selecting a single column without whitespace
data_discount = data.Discount
data_discount
```
